```python
# NGROK_AUTH_TOKEN: Đăng nhập https://dashboard.ngrok.com/get-started/your-authtoken để copy token

import os
from pyngrok import ngrok
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

NGROK_AUTH_TOKEN = os.getenv("NGROK_AUTH_TOKEN")
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Colab cell 3 — chạy streamlit (chạy nền) và mở tunnel
import time, subprocess, os
# bật streamlit (chạy nền, redirect log)
cmd = "streamlit run app.py --server.port 8501 --server.address 0.0.0.0 --server.headless true &> /content/streamlit.log &"
_ = subprocess.call(cmd, shell=True)
time.sleep(2)  # chờ vài giây để streamlit khởi động

# mở tunnel public tới port 8501
public_url = ngrok.connect(8501)
print("Open this URL in your browser →", public_url)
```
